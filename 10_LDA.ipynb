```python
import gensim
import pyLDAvis
from pyLDAvis import gensim as gensimvis
import spacy

from pprint import pprint

import pandas as pd
```

```python
df = pd.read_pickle('Data/clean_seattle')
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
        

clean_sents = list(sent_to_words(df.listed_items))
```

```python
id2word = gensim.corpora.Dictionary(clean_sents)

corpus = [id2word.doc2bow(t) for t in clean_sents]
```

```python
[(id2word[id], freq) for id, freq in corpus[0]]
```

```python
lda_model = gensim.models.ldamulticore.LdaMulticore(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=10, 
                                           random_state=100,
                                           chunksize=100,
                                           passes=128,
                                           per_word_topics=True)
```

```python
pprint(lda_model.print_topics())
```

```python
print(df.listed_items[0])
```

```python
top_topics = lda_model.get_document_topics(corpus[0])
top_topics.sort(key=lambda x: x[1], reverse=True)

print(top_topics)
```

```python
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, corpus, id2word)
pyLDAvis.save_html(vis, 'Visualizations/lda.html')
```
