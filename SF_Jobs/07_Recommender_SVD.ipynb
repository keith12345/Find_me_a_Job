```python
import pandas as pd

from sklearn.decomposition import PCA

import pickle
```

## Adding SVD

```python
df = pd.read_pickle('../Data/01_tf_idf_and_features')
```

```python
df.head(2)
```

```python
pca = PCA(n_components=100)
pca.fit(df.iloc[:,4:])
tf_idf_df = pd.DataFrame(pca.transform(df.iloc[:,4:]))
```

```python
pickle_out = open('../Tools_and_models/pca_tool',"wb")
pickle.dump(pca, pickle_out)
pickle_out.close()
```

```python
df = df.iloc[:,:4].merge(tf_idf_df,left_index=True,right_index=True)
```
