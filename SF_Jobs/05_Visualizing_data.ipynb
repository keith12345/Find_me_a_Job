```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle('../Data/01_sf_labeled')
```

```python
partition_size = pd.DataFrame(df.labels.value_counts()).reset_index()
partition_size.columns = ['labels','partition_size']
partition_size['partition_size'] = (partition_size
                                    .partition_size
                                    .apply(lambda x: 'Partition Size: ' + str(x)))
partition_size
```

```python
df = df.merge(partition_size,on='labels')
```

```python
df.head()
```

```python
df['job_title'] = df.job_title.apply(lambda x: 'Actual Title: ' + x)
```

```python
df['company_name'] = df.company_name.apply(lambda x: 'Company Name: ' + x)
```

```python
df['hover_info'] = (df.company_name + '<br>' 
                    + df.labels + '<br>' 
                    + df.job_title + '<br>' 
                    + df.partition_size + '<br>' 
                    + df.url)
```
